```python
#declaring dependent and independent features or variables
X = df.iloc[:, :-2]
y = iris.target
```

```python
print(X)
```

```python
print(X.shape, y.shape)
```

```python
#building the classification model
#importing the libraries
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
#splitting the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#checking the number of rows and columns
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
#training the model
treemodel = DecisionTreeClassifier()
treemodel.fit(X_train, y_train)
```

```python
#predicting the future class
ypred = treemodel.predict(X_test)
print(ypred)
```

```python
#visualizing the decision tree graphically
from sklearn import tree
plt.figure(figsize=(15, 10))
tree.plot_tree(treemodel, filled=True)
```

```python
#evaluating the model score
score = accuracy_score(ypred, y_test)
print(score)
print(classification_report(ypred, y_test))
```
